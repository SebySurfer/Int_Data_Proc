```python
import pandas as pd
from sklearn.linear_model import LinearRegression

path = "D_ConstructorInmobiliario.csv"
df = pd.read_csv(path)

df.head()
```

```python
# Identificar variables
Y_col = "SqrFt"  # variable dependiente
X_cols = ["Income", "Size", "Senior", "Education"]  # variables independientes

X = df[X_cols]
y = df[Y_col]

print("Variable dependiente (Y):", Y_col)
print("Variables independientes (X):", X_cols)
print("\nResumen de datos:")
print(df[[Y_col] + X_cols].describe())
```

```python
# Ajustar regresión lineal múltiple
model = LinearRegression()
model.fit(X, y)

b0 = float(model.intercept_)
b_income, b_size, b_senior, b_edu = [float(c) for c in model.coef_]

print("Coeficientes del modelo")
print(f"b0 (intercepto) = {b0:.6f}")
print(f"b1 (Income)    = {b_income:.6f}")
print(f"b2 (Size)      = {b_size:.6f}")
print(f"b3 (Senior)    = {b_senior:.6f}")
print(f"b4 (Education) = {b_edu:.6f}")

# Ecuación de regresión múltiple
equation = (
    f"SqrFt = {b0:.6f} + ({b_income:.6f})·Income + ({b_size:.6f})·Size "
    f"+ ({b_senior:.6f})·Senior + ({b_edu:.6f})·Education"
)
print("\nEcuación de regresión múltiple:")
print(equation)

print("\nR^2 (bondad de ajuste, en entrenamiento):", round(model.score(X, y), 6))
```
