```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
%matplotlib inline

from sklearn import linear_model
from sklearn.metrics import mean_squared_error, r2_score

```

```python
import pandas as pd

data = pd.read_csv("CopierSalesofAmerica.csv")
```

```python
data.shape
```

```python
data.head()
```

```python
#Seleccionar las características y la variable objetivo
x_train = np.array(data['llamadas']).reshape(-1,1)
y_train = np.array(data['ventas'])
```

```python
r = linear_model.LinearRegression()
r.fit(x_train, y_train)

# Hacer predicciones
y_pred = r.predict(x_train)

print('Coeficientes: ', r.coef_)
print('Intercepto: ', r.intercept_)
print('Error cuadrático medio: ', mean_squared_error(y_train, y_pred))
print('Varianza: ', r2_score(y_train, y_pred))

```
