```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
%matplotlib inline

from sklearn import linear_model
from sklearn.metrics import mean_squared_error, r2_score

```

```python
import pandas as pd

data = pd.read_csv("/Users/Sebastian 1/Desktop/IT Class/Classes/Int_Data_Proc/Act0_RegresionLineal/CopierSalesofAmerica.csv")
```

```python
data.shape
```

```python
data.head()
```

```python
#Seleccionar las características y la variable objetivo
x_train = np.array(data['llamadas']).reshape(-1,1)
y_train = np.array(data['ventas'])
```

```python
r = linear_model.LinearRegression()
r.fit(x_train, y_train)

# Hacer predicciones
y_pred = r.predict(x_train)

print('Coeficientes: ', r.coef_)
print('Intercepto: ', r.intercept_)
print('Error cuadrático medio: ', mean_squared_error(y_train, y_pred))
print('Varianza: ', r2_score(y_train, y_pred))

```

```python
x = x_train.ravel()
plt.figure(figsize=(10, 5))
plt.scatter(x, y_train, color='blue', label='Datos reales', s=80)
plt.plot(x, y_pred, color='red', label='Pendiente', linewidth=2)
plt.xlabel('No. de llamadas')
plt.ylabel('Ventas de copiadoras')
plt.title('Regresión Lineal - Copiadoras vendidas')
plt.legend()
plt.grid(True, linestyle='--', color='0.75')
plt.show()
```
