# 1.1 LA NATURALEZA COMO OPTTMIZADORA

En palabras resumidas, se redacta como la evolucion y la naturaleza llega hacerse la perfeccion en comparacion con algo hecho artificialmente. La vida como tal nos enseña que dentro de este mundo imperfecto, cada ser vivo ha llegado a muchos evoluciones para adaptarse a los cambios de su alrededor, siendo esto el factor mas importante de la sobrevivencia de una especie. En otras palabras, los especies no sobreviven por su inteligencia o que tan fuertes son, pero la adaptacion del mismo influye completamente en su resultado final.

Siendo que queremos imitar la inteligencia, deberiamos igualmente imitar la naturaleza. Si no es asi, no llegariamos a un inteligencia superior al nuestro. 



# 1.2 UN POCO DE BIOLOGIA

Los seres vivos tenemos aspectos fisicos que nos diferencian entre nosotros, siendo los fenotipos. Todo por esto resulta de parte genetico de las cromosomas que se heredan por generaciones, de padres a hijos. Por cada generacion, la mitad de la genetica de uno se une con el otro, creando un nuevo ser vivo. Esto genera un nuevo vinculo de posibilidad con las genomas que an resultado la sobrevivencia de estas mismas generaciones. Por lo tanto, no siempre se mantiene igual, y puede ocasionar cambios inesperados, siendo la mutuacion del genoma. Con gran frecuencia, esto genera fenotipos mortales al individuo, siendo que un codigo usado por millones de años ha servido con buenas razones. Por lo tanto, hay ocasiones pequeñas y especiales, donde este cambio de "ritmo" pueda beneficiar el ser vivo, y hasta escribir todo una nueva generacion por completa por el hecho de mejorar su sobrevivencia.
