# 1.2 UN POCO DE BIOLOGIA

Los seres vivos tenemos aspectos fisicos que nos diferencian entre nosotros, siendo los fenotipos. Todo por esto resulta de parte genetico de las cromosomas que se heredan por generaciones, de padres a hijos. Por cada generacion, la mitad de la genetica de uno se une con el otro, creando un nuevo ser vivo. Esto genera un nuevo vinculo de posibilidad con las genomas que an resultado la sobrevivencia de estas mismas generaciones. Por lo tanto, no siempre se mantiene igual, y puede ocasionar cambios inesperados, siendo la mutuacion del genoma. Con gran frecuencia, esto genera fenotipos mortales al individuo, siendo que un codigo usado por millones de años ha servido con buenas razones. Por lo tanto, hay ocasiones pequeñas y especiales, donde este cambio de "ritmo" pueda beneficiar el ser vivo, y hasta escribir todo una nueva generacion por completa por el hecho de mejorar su sobrevivencia. 

Al ya aplicarlo en programacion y algoritmos genericos, inicialmente se manifiesta una gran gama de codigo que podra ser util ante cualquier situacion. Ya dado un problema de resolver, se adapta comforme se resuelva el problema, y elimina lo inecesario. Tambien se podria modificar con optimizaciones un codigo existente que resuelve el problema o se aproxime a resolverlo mejor. 

Se supone que tenemos que generar muchas posibles soluciones aleatorias, despues seleccionar cual es el "menos malo", y al final re-iterar otra generacion con el aprendizaje. En mi opiniion, siento yo que esto NO deberia de hacer un modelo apadtable. En realidad, es mejor tratar de hacer los mejores modelos posibles para resolver un problema, y cada uno de ellos teniendo un bias completamente diferente. Ya al averiguar cual de todos de estos modelos hacen la mejor solucion, se brinca al mejor solucion posible de forma mas rapido, claridad y eficiencia. Aunque podra tomar mas tiempo de trabajo, brinda mejor calidad. 

```python
import random


class Individuo:
    def __init__(self, poder, velocidad, resistente_temperatura, inteligencia):
        self.masa = poder
        self.velocidad = velocidad
        self.resistente_temperatura = resistente_temperatura
        self.inteligencia = inteligencia


class Poblacion:
    def __init__(self, tipo_ambiente, tipo_individuo):
        self.tipo_ambiente = tipo_ambiente
        self.tipo_individuo = tipo_individuo

    def generar_individuos(self):
        individuos = []
        combinaciones_usadas = set()

        while len(individuos) < 100:
            valores = (
                random.randint(1, 10),  # poder
                random.randint(1, 10),  # velocidad
                random.randint(1, 10),  # resistente_temperatura
                random.randint(1, 10),  # inteligencia
            )

            if valores not in combinaciones_usadas:
                combinaciones_usadas.add(valores)
                individuos.append(Individuo(*valores))

        return individuos
```
