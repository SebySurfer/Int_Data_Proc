```python
import pandas as pd
from sklearn.linear_model import LinearRegression

# Cargar datos
path = "C_Coleccionista.csv"
df = pd.read_csv(path)

df.head()
```

```python
# Identificar variables
Y_col = "Price"                 # variable dependiente
X_cols = ["Bidders", "Age"]     # variables independientes

X = df[X_cols]
y = df[Y_col]

print("Variable dependiente (Y):", Y_col)
print("Variables independientes (X):", X_cols)
print("\nResumen de datos:")
print(df[[Y_col] + X_cols].describe())
```

```python
# Ajustar regresión lineal múltiple
model = LinearRegression()
model.fit(X, y)

b0 = float(model.intercept_)
b1, b2 = [float(c) for c in model.coef_]

print("Coeficientes del modelo")
print(f"b0 (intercepto) = {b0:.6f}")
print(f"b1 (Bidders)    = {b1:.6f}")
print(f"b2 (Age)        = {b2:.6f}")

# Ecuación de regresión múltiple
equation = (
    f"Price = {b0:.6f} + ({b1:.6f})·Bidders + ({b2:.6f})·Age"
)
print("\nEcuación de regresión múltiple:")
print(equation)

print("\nR^2 (bondad de ajuste, en entrenamiento):", round(model.score(X, y), 6))
```
