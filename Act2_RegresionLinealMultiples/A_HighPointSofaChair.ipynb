```python
import pandas as pd
from pathlib import Path

path1 = Path("Act2_RegresionLinealMultiple") / "A_HighPointSofaandChair.csv"
path2 = Path("A_HighPointSofaandChair.csv")

csv_path = path1 if path1.exists() else path2

df = pd.read_csv(csv_path)
df
```

```python
# Definir variables (y = dependiente, X = independientes)

y_col = "Performance"
x_cols = ["Aptitude", "Union"]

y = df[y_col]
X = df[x_cols]

print("Variable dependiente (y):", y_col)
print("Variables independientes (X):", ", ".join(x_cols))

# Ajuste del modelo (statsmodels si está disponible; si no, sklearn)
try:
    import statsmodels.api as sm
    from IPython.display import display, Math

    X_sm = sm.add_constant(X)
    model = sm.OLS(y, X_sm).fit()

    display(model.summary())

    b0 = model.params["const"]
    b1 = model.params[x_cols[0]]
    b2 = model.params[x_cols[1]]

    display(Math(
        rf"\hat{{{y_col}}} = {b0:.4f} + ({b1:.4f})\,{x_cols[0]} + ({b2:.4f})\,{x_cols[1]}"
    ))
except Exception:
    from sklearn.linear_model import LinearRegression

    lr = LinearRegression()
    lr.fit(X, y)

    b0 = lr.intercept_
    b1, b2 = lr.coef_

    print("\nEcuación de regresión múltiple (estimada):")
    print(f"{y_col}^ = {b0:.4f} + ({b1:.4f})*{x_cols[0]} + ({b2:.4f})*{x_cols[1]}")
```
